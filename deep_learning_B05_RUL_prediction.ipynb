```python
## Data 
filenames = 'B0005_data_x_train.csv'
x_train = np.loadtxt('B0005_data_x_train.csv', delimiter=',')
filenames = 'B005_data_y_train.csv'
y_train = np.loadtxt('B005_data_y_train.csv', delimiter=',')

filenames = 'B0005_data_x_test.csv'
x_test = np.loadtxt('B0005_data_x_test.csv', delimiter=',')

filenames = 'B005_data_y_test.csv'
y_test = np.loadtxt('B005_data_y_test.csv', delimiter=',')

x_train = np.array(x_train,dtype =float)
y_train = np.array(y_train,dtype =float)
x_test = np.array(x_test,dtype =float)
y_test = np.array(y_test,dtype =float)

x_train = x_train.reshape(100,1,2,3)
x_test = x_test.reshape(40,1,2,3)
y_train.shape
```

```python
model= Sequential()
model.add(Convolution2D(
    batch_input_shape=(None, 1, 2, 3),
    kernel_size = (3,3),
    filters=3,
    strides=2,
    padding='same',      # Padding method
    data_format='channels_first',
))

# Convolutional layer: 96 11*11 lters with a stride of 4 # 1
model.add(Convolution2D(
    kernel_size = (3,3),
    filters=2,
    strides=2,
    padding='same',      # Padding method
    data_format='channels_first',
))

# # Max pooling: 3*3 max pooling with a stride of 2 # 2
model.add(MaxPooling2D(
    pool_size=(3,3),
    strides=2,
    padding='same',    # Padding method
    data_format='channels_first',
))
# model.add(BatchNormalization()) # Batchnormalization 1
#  model.add(Activation('relu'))

model.add(Flatten())
model.add(Dense((8),batch_input_shape=(None,6)))
model.add(Dense((8),batch_input_shape=(None,6)))
model.add(Dense((8),batch_input_shape=(None,6)))
model.add(Dense((8),batch_input_shape=(None,6)))
model.add(Dense((1),batch_input_shape=(None,None)))
model.compile(loss = 'mean_absolute_error',optimizer = 'adam',metrics=['accuracy'])
model.summary()
```

```python
history2 = model.fit(x_train,y_train,epochs=2000,validation_data=(x_test,y_test))
```

```python
# Loss: convergence or not
plt.plot(history2.history['loss'])
plt.show()
```

```python
# this function is to show the location in the array of first value which reach its failure capacity
# y_failure = failure_capacity*np.ones(40).reshape(40)
def find_the_failed_cycle(predicted_results,failed_capacity):
    for i in range(predicted_results.shape[0]):
        if(predicted_results[i]<failed_capacity):
            print(i)
            break # when find, then pause and exit
    return i
```

```python
failed_capacity = 1.4
# actual RUL
print('Actual RUL ')
actual_RUL = find_the_failed_cycle(actual,failed_capacity)
# DNN
print('DNN RUL ')
DNN_RUL = find_the_failed_cycle(results_DNN1,failed_capacity)
DNN_error = abs(DNN_RUL-actual_RUL)
#LSTM_DNN
print('LSTM_DNN RUL ')
LSTM_RUL = find_the_failed_cycle(results_LSTM1,failed_capacity)
LSTM_error = abs(LSTM_RUL-actual_RUL)
#GRU_DNN
print('GRU_DNN RUL ')
GRU_RUL = find_the_failed_cycle(results_GRU1,failed_capacity)
GRU_error = abs(GRU_RUL-actual_RUL)
#CNN_DNN
print('CNN RUL ')
CNN_RUL = find_the_failed_cycle(results_CNN1,failed_capacity)
CNN_error = abs(CNN_RUL -actual_RUL)
#CNN_LSTM_RUL
print('CNN_LSTM_DNN RUL ')
CNN_LSTM_RUL = find_the_failed_cycle(results_CNN_LSTM1,failed_capacity)
CNN_LSTM_error = abs(CNN_LSTM_RUL-actual_RUL)
#CNN_GRU_RUL
print('CNN_GRU_DNN RUL ')
CNN_GRU_RUL = find_the_failed_cycle(results_CNN_GRU1,failed_capacity)
CNN_GRU_error = abs(CNN_GRU_RUL-actual_RUL)

# plot bar_figure
plt.figure(figsize=(6,3))
target = ['LSTM_DNN','GRU_DNN','DNN','CNN_DNN','CNN_LSTM_DNN','CNN_GRU_DNN']
value = (LSTM_error,GRU_error,DNN_error,CNN_error,CNN_LSTM_error,CNN_GRU_error)
plt.ylabel('Prediction Error')
plt.bar(target,value,width = 0.8, color = 'r')
# plt.xticks(fontsize = 7)
plt.xticks(rotation = 15)
plt.title('Prediction Error')
plt.show()
```
